# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np 


QUERY = """
SELECT Respondent, ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

salaries = df['ConvertedComp'].dropna()

def format_fn(value, tick_number):
    return f'{int(value)}'



count, bin_edges = np.histogram(salaries)

df['ConvertedComp'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.gca().xaxis.set_major_formatter(FuncFormatter(format_fn))

plt.xticks(rotation=45)

plt.title('Histogram of Salaries') # add a title to the histogram
plt.ylabel('Number of Respondents') # add y-label
plt.xlabel('Compensation') # add x-label

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT *
FROM master
"""
df = pd.read_sql_query(QUERY, conn)

df.head()

df['Age'].plot(kind='box', figsize=(8, 6))

plt.title('Age')
plt.ylabel('age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

#df = df[df['WorkWeekHrs'] < 100]

df.plot(kind='scatter', x='WorkWeekHrs', y='Age', figsize=(10, 6), color='darkblue')

plt.title('working hours by age')
plt.xlabel('Working Hours')
plt.ylabel('Age')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
ax0 = df.plot(kind='scatter',
                    x='WorkWeekHrs',
                    y='CodeRevHrs',
                    figsize=(14, 8),
                    alpha=0.5,  # transparency
                    color='green',
                    s=df['Age'] * 20  # pass in weights 
                    
                    )

plt.title('Code Reviews by Working Hours & age')
plt.xlabel('Working Hours')
plt.ylabel('Code Rev Hours')

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT DatabaseDesireNextYear, count(*) as num
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY num desc
limit 5
"""
df = pd.read_sql_query(QUERY, conn)

df.head()

df['num'].plot(kind='pie',
                            figsize=(10, 6),
                            autopct='%1.1f%%', 
                            startangle=90,         
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.10,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            # colors=colors_list,  # add custom colors
                            # explode=explode_list # 'explode' lowest 3 continents
                            )

# scale the title up by 12% to match pctdistance
plt.title('Most Desired to Learn Databases', fontsize = 15) 

plt.axis('equal') 

# add legend
plt.legend(labels=df['DatabaseDesireNextYear'], loc='upper left', fontsize=7) 

plt.show()






```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, CodeRevHrs, WorkWeekHrs  FROM Master
WHERE Age BETWEEN 30 AND 35
ORDER BY Age
"""
df = pd.read_sql_query(QUERY, conn)

df_plot = df.groupby('Age').median()

df_plot.head()


# # # stacked Histogram
df_plot.plot(kind='bar',      
          stacked=True,
          color=['coral', 'darkslateblue'],
          
        
         )

plt.title('Median of CodeRev and WorkHours by Age')
plt.ylabel('Hours')
plt.xlabel('Age') 


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp  FROM Master
WHERE Age BETWEEN 45 AND 60
ORDER BY Age
"""
df = pd.read_sql_query(QUERY, conn)

df_plot = df.groupby('Age').median()

df_plot.head()

df_plot.plot(kind='line', figsize=(14, 8))

plt.title('Median Compensation by Age')
plt.ylabel('Compensation')
plt.xlabel('Age')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

import textwrap

QUERY = """
SELECT MainBranch, count(*)as count FROM Master
GROUP BY MainBranch

"""
df = pd.read_sql_query(QUERY, conn)

df = df.set_index('MainBranch')

wrapped_y_labels = [textwrap.fill(label, width=10) for label in df.index]

df.plot(kind='barh', figsize=(10, 6))

plt.title('Median Compensation by Age')
plt.ylabel('Main Branch')
plt.xlabel('Count')

plt.show()

df.head()
```
